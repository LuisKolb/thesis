```python
# sort dataframe and create a column that is the rank of the R5 score
import pandas as pd

df = pd.read_csv("df_retrieval.csv", index_col=0)
df.columns = ["R5", "MAP", "retriever", "dataset", "wall time (s)"]
df = df.sort_values(["R5"], ascending=False)
last_r5 = 0
rank = 0
rank_list = []
for row in df.values:
    if not row[1] == last_r5: rank += 1
    last_r5 = row[1]
    rank_list.append(rank)

df["rank"] = rank_list
cols = df.columns.tolist()
df = df[cols[-1:] + cols[:-1]] #reorder columns, first col to start
display(df.style.hide()) # .style.hide() hides the index column
```
